```python
import os
import cv2

# Path to your video file
video_path = '../video/0.45_crop.avi'

# Create a folder to save extracted frames
output_folder = '../video/Frames0.45_crop/'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Create a VideoCapture object
vidcap = cv2.VideoCapture(video_path)

# Check if the video is successfully opened
if not vidcap.isOpened():
    print("Error: Video could not be opened.")
    exit()

# Custom frame rate and duration
custom_frame_rate = 4600  # frames per second
custom_duration = 10  # seconds

# Calculate total frames to be extracted
total_frames = int(custom_frame_rate * custom_duration)

# Initialize variable to count extracted frames
count = 0

# Loop to extract frames
while count < total_frames:
    # Read frame
    success, image = vidcap.read()
    
    # Check if reading operation is successful
    if not success:
        print("Error: Failed to read frame.")
        break
    
    # Save frame as JPEG file in the output folder
    frame_filename = os.path.join(output_folder, "frame%d.bmp" % count)
    cv2.imwrite(frame_filename, image)
    
    # Add delay according to custom frame rate
    delay = int(1000 / custom_frame_rate)
    cv2.waitKey(delay)
    
    # Print frame extraction status
    print('Frame %d successfully extracted.' % count)
    
    # Increment frame count
    count += 1

# Release VideoCapture object
vidcap.release()
```
