```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Data Metode dan Manual
data_metode = np.array([1.27, 1.46, 1.46, 1.46, 1.44, 1.44, 1.47, 1.50, 1.48, 1.46, 1.46, 1.49, 1.51, 1.46, 1.50, 1.46, 1.42, 1.43, 1.45])
data_manual = np.array([1.70, 1.58, 1.65, 1.47, 2.16, 1.23, 2.30, 1.83, 1.18, 1.28, 1.87, 1.73, 1.98, 1.01])

# Pastikan panjang data sama dengan memangkas yang lebih panjang
min_length = min(len(data_metode), len(data_manual))
data_metode = data_metode[:min_length]
data_manual = data_manual[:min_length]

# Statistik Deskriptif
mean_metode = np.mean(data_metode)
mean_manual = np.mean(data_manual)
std_metode = np.std(data_metode)
std_manual = np.std(data_manual)

# Perbedaan
difference = data_metode - data_manual

# Uji Statistik (paired t-test)
t_stat, p_value = ttest_rel(data_metode, data_manual)

# Kesalahan
mae = np.mean(np.abs(difference))
mse = np.mean(difference**2)
rmse = np.sqrt(mse)

# Output Statistik
print(f"Mean Metode: {mean_metode}, Mean Manual: {mean_manual}")
print(f"Std Dev Metode: {std_metode}, Std Dev Manual: {std_manual}")
print(f"MAE: {mae}, MSE: {mse}, RMSE: {rmse}")
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Plot Data
plt.figure(figsize=(12, 6))

# Scatter Plot
plt.subplot(1, 3, 1)
plt.scatter(data_manual, data_metode)
plt.xlabel('Data Manual')
plt.ylabel('Data Metode')
plt.title('Scatter Plot')

# Line Plot
plt.subplot(1, 3, 2)
plt.plot(data_manual, label='Manual', marker='o')
plt.plot(data_metode, label='Metode', marker='o')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.title('Line Plot')

# Box Plot
plt.subplot(1, 3, 3)
plt.boxplot([data_manual, data_metode], labels=['Manual', 'Metode'])
plt.title('Box Plot')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Contoh data
nilai_manual = np.array([1.68, 1.70, 1.58, 1.65, 1.47, 2.16, 1.23, 2.30, 1.83, 1.18, 1.28, 1.87, 1.73, 1.98, 1.01, 2.0, 1.7, 1.8, 1.6, 1.7])
nilai_farneback = np.array([1.39, 1.60, 1.60, 1.60, 1.58, 1.59, 1.62, 1.66, 1.63, 1.59, 1.60, 1.63, 1.65, 1.60, 1.64, 1.59, 1.55, 1.57, 1.59, 1.6])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(nilai_manual.reshape(-1, 1), nilai_farneback, test_size=0.2, random_state=42)

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R Score: {r2}")
```
